```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# some function to be applied to the input array
f = lambda x: x**3-x**2

# input array consisting of 100 random numbers between 0 and 1
x = np.random.rand(100)
# generate target array by applying the function to the input array
t = f(x)

# plot the data points
plt.plot(x,t,'o',label='data points')

# plot the underlying function
x2 = np.linspace(0,1,100)
t2 = f(x2)
plt.plot(x2,t2,'k',label='underlying function')

plt.legend()
plt.show()
```
