```python
"""
@author: faurand, chardes, ehagensieker
"""
import numpy as np
import matplotlib.pyplot as plt
from mlp import MLP
from layer import Layer

#creating the Multilayer perceptron (1 hidden layer with 10 units)
mlp = MLP()
mlp.add_layer(Layer(10,1))
mlp.add_layer(Layer(1,10))

#building the data set
x = np.random.rand(100)
t = np.zeros(100)
for i,entry in enumerate(x):
  t[i] = entry**3 - entry**2

#array for saving the loss values
loss_average = []

for epoch in range(10):
  loss_list = []
  
  #training the network
  for i,y in zip(x,t):
  
    activation = mlp.forward_step(np.array([[i]]))
    loss = 1/2 * (activation-y)**2
    loss_derived = activation-y
    mlp.backpropagation(loss_derived,0.01)
    loss_list.append(loss)
    
  loss_average.append(np.mean(loss_list))
  print("loss: ",loss_average)

#visualiye the training progress
plt.plot(loss_average[:5])
plt.show


```
